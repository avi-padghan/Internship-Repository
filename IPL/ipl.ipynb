```python
# Highest Wicket Taker 
wickets = deliveries[(deliveries['is_wicket']== 1)
                    & (deliveries['dismissal_kind']!= 'run out')]
```

```python
highest_wicket_taker = wickets.head(1)
print('Highest Wicket Taker',highest_wicket_taker)
```

```python
top_wickets_taker =wickets.head(5)
print(top_wickets_taker)
```

```python
# Toss Impact 
matches['Toss_Match_Win'] = matches['toss_winner'] == matches['winner']
toss = matches['Toss_Match_Win'].value_counts()
toss.head(5)
```

```python
plt.figure(figsize=(5,5))

plt.pie(
    toss.values,
    labels=['Won Match', 'Lost Match'],
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Impact of Toss on Match Result")

plt.show()
```

```python
orange_cap = (
    merge_df.groupby(['season','batter'])['batsman_runs']
    .sum()
    .reset_index()
)
```

```python
idx = orange_cap.groupby('season')['batsman_runs'].idxmax()

orange_cap = orange_cap.loc[idx]

print(orange_cap)
```

```python
wickets_merge = merge_df[
    (merge_df['is_wicket']==1) &
    (merge_df['dismissal_kind']!='run out')
]
```
