```python
sns.boxplot(x=df['Sales'])
plt.show()
```

```python
# Total Sales By Category

sales_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.Figure(figsize=(8,6))
plt.bar(sales_category.index,sales_category.values,)
plt.title('Total Sales by Category')
plt.ylabel('Sales')
plt.xlabel('Category')
plt.tight_layout()
plt.show()
```


Technology has the highest sales.
Furniture has good sales, but less than Technology.
Office Supplies has the lowest sales.
Technology is the best-selling category.
The sales of Furniture and Office Supplies are close to each other.


```python
# Sales by Region
sales_region = df.groupby('Region')['Sales'].sum()
plt.pie(sales_region.values,labels=sales_region.index,autopct='%1.1f%%',startangle=90)
plt.title('Sales Distribution by Region')
plt.tight_layout()
plt.show()
```

The West region has the highest sales (31.8%).
The East region has the second highest sales (29.7%).
The Central region has 21.6% of total sales.
The South region has the lowest sales (16.8%).
The company sells the most in the West region and the least in the South region.

```python
# Sales Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Sales'],bins=30)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Most sales are small amounts.
Only a few sales are very large.
The sales are not evenly distributed.
Most customers make low-value purchases.
High-value sales happen less often

```python
# top 10 Product Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.barh(top_products.index,
         top_products.values)

plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")

plt.show()
```

Canon imageCLASS 2200 Advanced Copier has the highest sales.
Fellowes PB500 is the second highest-selling product.
The other products have much lower sales.
One product sells much more than the others.

```python
#Monthly Sales Trend
df['Order Date'] = pd.to_datetime(df['Order Date'],format='mixed',errors='coerce')
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,5))

plt.plot(monthly_sales.index.astype(str),
         monthly_sales.values,
         marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

Sales go up and down every month.
Some months have very high sales.
The highest sales are at the end of 2026.
Overall, sales show an increasing trend over time.

Higher sales usually lead to higher profit.
Most sales have small profits.
Some orders have high sales but low or negative profit.
There are a few unusual (outlier) sales with very high or very low profit.
